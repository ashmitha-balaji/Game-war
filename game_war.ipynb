```python
##Card class

##suit , rank , value
suits = ('Hearts','Clubs','Spades','Diamond')
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen', 'King','Ace')
values = {'Two':2 ,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':11,'Queen':12, 'King':13,'Ace':14}
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python

#class deck ----create a class with all cards in it ---- 52 cards deck
```

```python
import random
class Deck():
    
    def __init__(self):
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                #create card obj
                card_created = Card(suit,rank)
                
                self.all_cards.append(card_created)
     
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
    def split_cards(self):
        split_cards = len(self.all_cards)//2
        return self.all_cards[:split_cards], self.all_cards[split_cards:]
    
    def one_card(self):
        
        return self.all_cards.pop()
    
        
    def get_all_cards(self):
        
        return self.all_cards
                
```

```python
class Player():
    def __init__(self,name):
        
        self.name = name
        self.all_cards = []
        
        
    def card_draw(self):
        return self.all_cards.pop(0)
    
    def clear_deck(self):
        return self.all_cards.clear()
    
    # add one card or list of cards
    def add_cards(self,new_cards):
        if type(new_cards) == type([]):
            self.all_cards.extend(new_cards)
            
        else:
            self.all_cards.append(new_cards)
            
    def __str__(self):
        return f'Player {self.name} has {len(self.all_cards)} cards'
```

```python
def game_setup():
    new_deck = Deck()
    new_deck.shuffle()
    split_list1,split_list2 = new_deck.split_cards()
#     print('list1')
#     for card in split_list1:
#         print(card)
#     print('list2')
#     for card in split_list2:
#         print(card)
    Player1 = Player('Roop')
    Player2 = Player('Kumar')
    Player1.add_cards(split_list1)
    Player2.add_cards(split_list2)
    return Player1,Player2

def game_begin():
    Player1, Player2 = game_setup()
    temp_deck = []
    rounds = 1
    
    while len(Player1.all_cards) != 0 and len(Player2.all_cards) != 0 and rounds <= 500:
        player1_card = Player1.card_draw()
        player2_card = Player2.card_draw()
        print(f'round no : {rounds}')
        print(f'Player1 card is {player1_card}')
        print(f'Player2 card is {player2_card}')
        rounds += 1
        if player1_card.value > player2_card.value:
            Player1.add_cards(player1_card)
            Player1.add_cards(player2_card)
            Player1.add_cards(temp_deck)
            temp_deck.clear()
            if len(Player2.all_cards)== 0:
                print('Congrats! Player1 won.')
                return
            
        elif player2_card.value > player1_card.value:
            Player2.add_cards(player1_card)
            Player2.add_cards(player2_card)
            Player2.add_cards(temp_deck)
            temp_deck.clear()
            if len(Player1.all_cards)== 0:
                print('Congrats! Player2 won.')
                return
            
        elif player1_card.value == player2_card.value:
            if len(Player1.all_cards)<= 15:
                print(f"Player {Player1.name} doesn't have enough cards. Player {Player2.name} wins!! ")
                return
            elif len(Player2.all_cards)<= 15:
                print(f"Player {Player2.name} doesn't have enough cards. Player {Player1.name} wins!! ")
                return
            
            else:
                temp_deck.append(player1_card)
                temp_deck.append(player2_card)
                continue
            
        
    print('Game Draw')
    
    
```
